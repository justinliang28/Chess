```python
import random
```

```python
def game(weights=[1,1,1,1,1,1],trials=10):
    List = []
    for i in range(len(weights)):
        for j in range(weights[i]):
            List.append(i + 1)
    outcome = []
    for i in range(trials):
        outcome.append(random.choice(List))
    
    length = len(outcome)
    experiment_weights = []
    for i in range(6):
        experiment_weights.append(outcome.count(i+1)/length)
    return outcome,experiment_weights
```

```python
outcome,experiment_weights = game(weights = [1,2,1,2,1,2],trials = 100000)
```
